```python
import numpy as np
import matplotlib.pyplot as plt
```

# Logistic regression

```python
def f(x, beta_0, beta):
    return 1 / (1 + np.exp(-(beta_0 + x * beta)))
```

```python
beta_0 = 0
beta_seq = np.array([0,0.2,0.5,1,2,10])
x = np.arange(-10, 10, 0.1)

plt.xlabel("x")
plt.ylabel("P(Y=1|x)")
plt.title("logistic curve")

for i in range(0, len(beta_seq)):
    beta = beta_seq[i]
    p = f(x, beta_0, beta)
    plt.plot(x, p, label=f'{beta}')

plt.legend(loc='upper left')
```
